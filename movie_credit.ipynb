```python
import ast# 
```

```python
def convert(obj):
  l=[]
  for t in ast.literal_eval(obj): # it goes to string and consider inside data 
    l.append(t['name'])
  return l
```

```python
movie.genres = movie.genres.apply(convert)
movie
```

```python
movie.iloc[0].keywords
```

```python
movie.keywords = movie.keywords.apply(convert)
movie
```

```python
def convert_cast(obj):
  l=[]
  counter = 0 
  for t in ast.literal_eval(obj): # it goes to string and consider inside data 
    if counter !=3: # we take only top 3 cast otherwise you take all of them depends task to task
      l.append(t['name'])
      counter+=1
    else:
      break
  return l
```

```python
movie.cast = movie.cast.apply(convert_cast)
movie.cast
```

```python
movie.crew[0]
```

```python
def get_director(obj):
  l=[]
  for t in ast.literal_eval(obj): # it goes to string and consider inside data 
    if (t['job']=='Director'):
      l.append(t['name'])
    break # only one director 
  return l
```

```python
movie.crew = movie.crew.apply(get_director)
movie.crew
```

# Stemming

```python
import nltk
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'][0]
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Similarity Check

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# need to get top 5 high sililarity scores
# we can sort it in descending order
sorted(similarity[0],reverse=True)
#This will sort the series but index postion will also be moved
# we will not only sort but also generate new indices using enumerate
```

```python
# this will sort on new indices 
# sorted(list(enumerate(similarity[0])))
#this will held the indexes also 
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x: x[1])[1:10]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]]['title'])
```

```python
recommend('Avatar')
```

```python
recommend("The Dark Knight Rises")
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
